#### 3. 单因子,双因子下个性因子分析
* 单因子误识下, 个性因子中, 分辨率,cpu核心,cpu kernel,系统指纹这四个属性对设备找回错误的影响最小. (即若这4个属性相同, 很少会导致误识的情况), 其余属性对误识的影响几乎相等
* 双因子的情况与单因子一样, 个性因子中分辨率,cpu核心,cpu kernel,系统指纹这四个属性相对稳定, 不容易产生误识

```python
counts = [5183112,5182822,5181927,5175200,5163140,5162231,5161963,5104646,5094807,3777899,3423527,2538002,485468]
_index = ['manufacturer','cpu_core_number','brand','hardware','equipment','product','model','mainboard','size','resolution_ratio','cpu_kernel','cpu_core','system_fingerprint']
data = pd.Series(counts,index=_index)
sort_data = data.sort_values(ascending=False)
bar = Bar('单因子真误识','')

bar.add('',sort_data.index,sort_data.values)

percentage = sort_data/data.sum()
percentage = percentage.round(2)
bar2 = Bar('单因子真误识','')
bar2.add('',percentage.index,percentage.values)

grid = Grid()
grid.add(bar2, grid_left="60%")
grid.add(bar, grid_right="60%")
grid

percentage
```

```python
counts = [5964662,5965744,5963789,5961118,5955835,5955359,5954943,5924552,5903337,4307307,4573133,2071709,5550911]
_index = ['manufacturer','cpu_core_number','brand','hardware','equipment','product','model','mainboard','size','resolution_ratio','cpu_kernel','cpu_core','system_fingerprint']
data = pd.Series(counts,index=_index)
sort_data = data.sort_values(ascending=False)
bar = Bar('双因子真误识','')
bar.add('',sort_data.index,sort_data.values)
percentage = sort_data/data.sum()
percentage = percentage.round(2)
bar2 = Bar('双因子真误识','')
bar2.add('',percentage.index,percentage.values)

grid = Grid()
grid.add(bar2, grid_left="60%")
grid.add(bar, grid_right="60%")
grid

percentage
```

* 核心因子更容易导致拒识的情况, 而对于个性因子对拒识的影响, 有明显梯度趋势. 其中, 系统指纹,model,product,设备号, cpu核心个数,主板型号这几个属性在拒识设备中取相同值的占比最小. 说明拒识设备中, 一个magic对应多的xid组中, 不同xid在这些属性上的取值不同. 而这几个属性多为id属性, 本身重合的概率不大, 因此可能是2个:
    * magic的准确度不高, 即magic受这些属性的影响很小
    * 这些属性被篡改过, 但magic本身并不依赖这些属性计算

```python
counts = [33924,38588,28913,12922,2863,2416,2383,4346,21433,8563,10870,3859,1900]
_index = ['manufacturer','cpu_core_number','brand','hardware','equipment','product','model','mainboard','size','resolution_ratio','cpu_kernel','cpu_core','system_fingerprint']
data = pd.Series(counts,index=_index)
sort_data = data.sort_values(ascending=False)
bar = Bar('单因子真拒识','')
bar.add('',sort_data.index,sort_data.values)

percentage = sort_data/data.sum()
percentage = percentage.round(2)
bar2 = Bar('单因子真拒识','')
bar2.add('',percentage.index,percentage.values)

grid = Grid()
grid.add(bar2, grid_left="60%")
grid.add(bar, grid_right="60%")
grid
```
